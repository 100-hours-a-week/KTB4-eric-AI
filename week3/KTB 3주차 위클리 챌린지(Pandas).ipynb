```python
# 그룹화 미니퀘스트 1번

data = {
    '이름': ['홍길동', '김철수', '박영희', '이순신'],
    '부서': ['영업', '영업', '인사', '인사'],
    '급여': [5000, 5500, 4800, 5100]
}

df = pd.DataFrame(data)

grouped = df.groupby('부서')['급여'].sum().reset_index

print(grouped)
```

```python
# 그룹화 미니퀘스트 2번

data = {
    '이름': ['홍길동', '김철수', '박영희', '이순신', '강감찬', '신사임당'],
    '부서': ['영업', '영업', '인사', '인사', 'IT', 'IT'],
    '급여': [5000, 5500, 4800, 5100, 6000, 6200]
}

df = pd.DataFrame(data)

grouped = df.groupby('부서')['급여'].agg(['sum','mean'])

print(grouped)
```

```python
# 그룹화 미니퀘스트 3번

data = {
    '이름': ['홍길동', '김철수', '박영희', '이순신', '강감찬', '신사임당'],
    '부서': ['영업', '영업', '인사', '인사', 'IT', 'IT'],
    '급여': [5000, 5500, 4800, 5100, 6000, 6200]
}

df = pd.DataFrame(data)

grouped = df.groupby('부서')['급여'].mean().reset_index()

# filtered_df = df.groupby('부서').filter(lambda x: x['급여'].mean() >= 5000)

print(grouped[grouped['급여'] >= 5000])
```

```python
# 병합 미니퀘스트 1번

df1 = pd.DataFrame({'고객ID': [1, 2, 3], '이름': ['홍길동', '김철수', '이영희']})
df2 = pd.DataFrame({'고객ID': [2, 3, 4], '구매액': [10000, 20000, 30000]})

merged_inner = pd.merge(df1, df2, on='고객ID', how='inner')

print(merged_inner)
```

```python
# 병합 미니퀘스트 2번

df1 = pd.DataFrame({'고객ID': [1, 2, 3], '이름': ['홍길동', '김철수', '이영희']})
df2 = pd.DataFrame({'고객ID': [2, 3, 4], '구매액': [15000, 25000, 35000]})

merged_left = pd.merge(df1, df2, on="고객ID", how='left')

print(merged_left)
```

```python
# 병합 미니퀘스트 3번

df1 = pd.DataFrame({
    '고객ID': [1, 2, 3],
    '도시': ['서울', '부산', '대전'],
    '구매액': [10000, 20000, 30000]
})

df2 = pd.DataFrame({
    '고객ID': [1, 2, 3],
    '도시': ['서울', '부산', '광주'],
    '구매액': [15000, 25000, 35000]
})

merged_suffix = pd.merge(df1, df2, on=['고객ID','도시'] , how='inner', suffixes=('_df1', '_df2'))

print(merged_suffix)
```

```python
# 피벗 미니퀘스트 1번

data = {
    '날짜': ['2024-01-01', '2024-01-01', '2024-01-02', '2024-01-02'],
    '제품': ['A', 'B', 'A', 'B'],
    '판매량': [100, 200, 150, 250]
}

df = pd.DataFrame(data)

df_pivot = df.pivot(index='날짜', columns='제품', values='판매량')

print(df_pivot)
```

```python
# 피벗 미니퀘스트 2번

data = {
    '카테고리': ['전자', '가전', '전자', '가전'],
    '제품': ['A', 'B', 'A', 'B'],
    '판매량': [100, 200, 150, 250]
}

df = pd.DataFrame(data)

df_pivot_table = df.pivot_table(index='카테고리', columns='제품', values='판매량', aggfunc='sum')

print(df_pivot_table)
```

```python
# 중복 제거 미니퀘스트 1번

data = {
    '이름': ['김철수', '이영희', '김철수', '박민수'],
    '나이': [25, 30, 25, 40],
    '성별': ['남', '여', '남', '남']
}

df = pd.DataFrame(data)

df_unique = df.drop_duplicates()

print(df_unique)
```

```python
# 중복 제거 미니퀘스트 2번

data = {
    '제품': ['노트북', '태블릿', '노트북', '스마트폰'],
    '가격': [1500000, 800000, 1500000, 1000000],
    '카테고리': ['전자기기', '전자기기', '전자기기', '전자기기']
}

df = pd.DataFrame(data)

df_unique = df.drop_duplicates(subset='제품')

print(df_unique)

df['중복여부'] = df.duplicated()

print(df)
```

```python
# 문자열 처리 미니퀘스트 3번

df = pd.DataFrame({"설명": ["빅데이터 분석", "데이터 과학", "머신 러닝", "딥 러닝"]})

df["약어"] = df["설명"].apply(lambda x: "".join([word[0] for word in x.split()]))

print(df)
```
